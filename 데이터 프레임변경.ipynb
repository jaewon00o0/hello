```python
import requests
import pandas as pd
from IPython.display import display
```

```python
url = "https://jsonplaceholder.typicode.com/users"
```

```python
response = requests.get(url)
data = response.json()
```

```python
df = pd.DataFrame(data)
```

```python
address_df = pd.json_normalize(df['address'])
```

```python
display(df)
display(address_df)
```
