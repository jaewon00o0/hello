과제 1: All products의 책 제목 1 ~ 20개를 데이터프레임으로 받기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/index.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
title_list = []
```

```python
for h3 in soup.find_all('h3'):
    a_tag = h3.find('a')
    title = a_tag.get('title')
    title_list.append(title)
```

```python
df_20_books = pd.DataFrame({'Book Title': title_list})
```

```python
print(df_20_books)
```

과제 3: 10개 카테고리에서 10개씩 읽어와 가장 비싼 책 10개 제시하기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
base_url = "https://books.toscrape.com/"
response = requests.get(base_url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
target_categories = [
    "Travel", "Mystery", "Historical Fiction", "Sequential Art",
    "Classics", "Philosophy", "Romance", "Womens Fiction",
    "Fiction", "Childrens"
]

category_links = {}
```

```python
side_categories = soup.find('div', class_='side_categories').find_all('a')
for a_tag in side_categories:
    cat_name = a_tag.text.strip()
    if cat_name in target_categories:
        category_links[cat_name] = base_url + a_tag.get('href')

book_data = []
```

```python
for cat_name, cat_url in category_links.items():
    res = requests.get(cat_url)
    cat_soup = BeautifulSoup(res.text, "html.parser")
```

```python
books = cat_soup.find_all('article', class_='product_pod')[:10]
```

```python
for book in books:
        title = book.find('h3').find('a').get('title')

        price_text = book.find('p', class_='price_color').text
        price_float = float(re.sub(r'[^\d.]', '', price_text))

        book_data.append({
            'Category': cat_name,
            'Title': title,
            'Price(£)': price_float
        })
```

```python
df_category_books = pd.DataFrame(book_data)
```

```python
df_sorted = df_category_books.sort_values(by='Price(£)', ascending=False)
```

```python
top_10_expensive = df_sorted.head(10)
```

```python
print(top_10_expensive.to_string(index=False))
```
