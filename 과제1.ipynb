```python
import requests
import os
import json
import pandas as pd
from IPython.display import Image, display
```

```python
print("🚀 [단계 1] API 데이터 호출을 시작합니다...")
url = "https://ll.thespacedevs.com/2.0.0/launch/upcoming"
response = requests.get(url)
```

```python
if response.status_code == 200:
    data = response.json()
    print("✅ API 데이터 호출 성공!")
else:
    print(f"❌ 데이터 호출 실패. 상태 코드: {response.status_code}")
    data = {}
```

```python
save_dir = "rocket_data"
os.makedirs(save_dir, exist_ok=True)
file_path = os.path.join(save_dir, "launches.json")
```

```python
with open(file_path, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
print(f"✅ 데이터 저장 완료: '{file_path}'")
```

```python
if 'results' in data:
    df = pd.DataFrame(data['results'])
else:
    df = pd.DataFrame(data)
```

```python
print("\n🌠 [단계 2] 이미지 폴더 저장 시작...")
image_dir = "rocket_images"
os.makedirs(image_dir, exist_ok=True)
```

```python
with open(file_path, "r", encoding="utf-8") as f:
    saved_data = json.load(f)
```

```python
launch_list = saved_data.get('results', [])
```

```python
for i, item in enumerate(launch_list):
    image_url = item.get('image')
```

```python
if not image_url:
        print(f"⚠️ [{i+1}/{len(launch_list)}] 이미지가 없습니다. (건너뜀)")
```

```python
filename = image_url.split('/')[-1]
save_path = os.path.join(image_dir, filename)
```

```python
try:
        img_response = requests.get(image_url, timeout=10)
        img_response.raise_for_status()

        with open(save_path, 'wb') as img_file:
            img_file.write(img_response.content)
        print(f"📸 [{i+1}/{len(launch_list)}] 다운로드 완료: {filename}")

except requests.exceptions.RequestException as e:
        print(f"❌ [{i+1}/{len(launch_list)}] 다운로드 실패: {e}")
```

```python
print("\n" + "="*50)
print("🚀 데이터프레임 표 & 로켓 사진 미리보기 (상위 5개)")
print("="*50)
```
