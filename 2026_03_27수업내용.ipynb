```python
import pandas as pd
import gradio as gr
import plotly.express as px
from google.colab import ai
```

```python
def load_data():
    file_names = [
        '서울특별시 공공자전거 이용정보(일별)_2510.csv',
        '서울특별시 공공자전거 이용정보(일별)_2511.csv',
        '서울특별시 공공자전거 이용정보(일별)_2512.csv'
    ]

    df_list = []
    for file in file_names:
        try:
            df = pd.read_csv(file, encoding='cp949')
            df_list.append(df)
        except FileNotFoundError:
            print(f"경고: '{file}' 파일을 찾을 수 없습니다. 파일명을 확인해주세요.")
            continue
        except UnicodeDecodeError:
            # cp949로 안될 경우 utf-8 시도
            try:
                df = pd.read_csv(file, encoding='utf-8')
                df_list.append(df)
            except Exception as e:
                print(f"경고: '{file}' 파일을 utf-8로 읽는 중 오류 발생: {e}")
                continue

    if not df_list:
        return pd.DataFrame()

    merged_df = pd.concat(df_list, ignore_index=True)

    if '대여일자' in merged_df.columns:
        merged_df['대여일자'] = pd.to_datetime(merged_df['대여일자'], errors='coerce'
    if '이용건수' in merged_df.columns:
        merged_df['이용건수'] = pd.to_numeric(merged_df['이용건수'], errors='coerce').fillna(0)

    return merged_df
```

```python
df = load_data()
```

```python
df = load_data()

def get_summary_stats():
    if df is None or df.empty:
        return "데이터를 불러오지 못했습니다. CSV 파일 위치를 확인하세요."

    total_records = len(df)
    total_usage = df['이용건수'].sum() if '이용건수' in df.columns else "N/A"
    date_range = f"{df['대여일자'].min().date()} ~ {df['대여일자'].max().date()}" if '대여일자' in df.columns else "N/A"

    summary = f"""
    ###  데이터 병합 요약
    * **총 데이터 건수:** {total_records:,} 건
    * **총 이용 건수:** {total_usage:,} 건
    * **데이터 수집 기간:** {date_range}
    """
    return summary
```

```python
def plot_daily_trend():
    if df is None or df.empty or '대여일자' not in df.columns or '이용건수' not in df.columns:
        return None

    daily_df = df.groupby('대여일자')['이용건수'].sum().reset_index()
    fig = px.line(daily_df, x='대여일자', y='이용건수', title=' 일별 공공자전거 총 이용건수 추이')
    return fig
```

```python
def plot_age_gender_stats():
    if df is None or df.empty:
        return None, None

    fig1 = None
    if '연령대코드' in df.columns and '이용건수' in df.columns:
        age_df = df.groupby('연령대코드')['이용건수'].sum().reset_index()
        fig1 = px.bar(age_df, x='연령대코드', y='이용건수', title=' 연령대별 누적 이용건수', color='연령대코드')

    fig2 = None
    if '성별' in df.columns and '이용건수' in df.columns:
        gender_df = df.groupby('성별')['이용건수'].sum().reset_index()
        fig2 = px.pie(gender_df, names='성별', values='이용건수', title=' 성별 이용 비율')

    return fig1, fig2
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("# 서울특별시 공공자전거(따릉이) 3개월 분석 대시보드")
    gr.Markdown("3개의 월별 CSV 데이터를 병합하여 만든 기본 통계분석 사이트입니다.")

    if df is None or df.empty:
        gr.Markdown("###  에러: CSV 파일 데이터를 불러오지 못했습니다. 코드가 있는 폴더에 파일이 있는지 확인해주세요.")
    else:
        with gr.Tabs():
            with gr.TabItem(" 기본 요약 및 데이터"):
                summary_output = gr.Markdown(get_summary_stats())
                gr.Markdown("###  병합된 데이터 미리보기 (상위 100개)")
                gr.Dataframe(df.head(100), interactive=False)

            with gr.TabItem(" 일별 이용 추이"):
                trend_plot = gr.Plot(plot_daily_trend())

            with gr.TabItem(" 연령/성별 분석"):
                with gr.Row():
                    age_plot = gr.Plot()
                    gender_plot = gr.Plot()
                app.load(plot_age_gender_stats, outputs=[age_plot, gender_plot])

    app.launch(share=False)
```

```python
import requests
import pandas as pd
import gradio as gr
from google.colab import ai
```

```python
def fetch_and_translate(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"

    params = {'userId': int(user_id)} if user_id else {}
    translated_text = "번역할 내용이 없습니다."
    original_text_display = ""

    try:
        response = requests.get(url, params=params)
        data = response.json()
        df = pd.DataFrame(data)
    except Exception as e:
        return pd.DataFrame(), f"데이터 호출 에러: {e}", translated_text

    if df.empty:
        return df, "해당 userId의 데이터가 없습니다.", translated_text

    first_row = df.iloc[0]
    original_text_display = f"Title: {first_row['title']}\nBody: {first_row['body']}"

    prompt = f"다음 영어 문장을 한국어로 자연스럽게 번역해줘:\n{original_text_display}"
    try:
        ai_response = ai.generate_text(prompt)
        translated_text = ai_response.text if hasattr(ai_response, 'text') else str(ai_response)
    except Exception as e:
        translated_text = f"번역 중 오류가 발생했습니다.\n코랩 환경인지 확인해주세요: {e}"

    return df, original_text_display, translated_text
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("##  게시글 실시간 필터링 및 1행 AI 번역기")
    gr.Markdown("7번 과제(userId 필터링)와 9번 과제(첫 행 번역)를 통합한 화면입니다.")
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("## 게시글 실시간 필터링 및 1행 AI 번역기")
    gr.Markdown("7번 과제(userId 필터링)와 9번 과제(첫 행 번역)를 통합한 화면입니다.")

    with gr.Row():
        user_id_input = gr.Number(label="User ID 입력", value=1, precision=0)

    with gr.Row():
        df_output = gr.Dataframe(label="필터링된 데이터프레임", interactive=False)

    gr.Markdown("### 첫 번째 행(1행) 번역 결과 (9번 과제 - Colab AI)")
    with gr.Row():
        original_output = gr.Textbox(label="원본 텍스트 (1행)", lines=5, interactive=False)
        translated_output = gr.Textbox(label="한국어 번역", lines=5, interactive=False)

    user_id_input.change(
        fn=fetch_and_translate,
        inputs=user_id_input,
        outputs=[df_output, original_output, translated_output],
    )

    app.launch(debug=True)
```

```python
import gradio as gr
import requests
```

```python
API_URL = "https://api-inference.huggingface.co/models/Salesforce/blip-image-captioning-large"
```

```python
def analyze_image_with_hf(image_filepath):
    """
    업로드된 이미지를 Hugging Face API로 보내서 분석 결과를 받아오는 함수
    """
    if image_filepath is None:
        return "이미지를 업로드해주세요."

    try:
        with open(image_filepath, "rb") as f:
            data = f.read()

        response = requests.post(API_URL, data=data)
        result = response.json()

        if isinstance(result, list) and 'generated_text' in result[0]:
            return result[0]['generated_text']
        elif 'error' in result:
            return f"API 로딩 중입니다. 잠시 후 다시 시도해주세요. (에러: {result['error']})"
        else:
            return str(result)

    except Exception as e:
        return f"분석 중 오류가 발생했습니다: {e}"
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("##  10번 과제: Hugging Face 멀티모달 LLM 테스트")
    gr.Markdown("Hugging Face의 `BLIP` 모델을 활용하여, 업로드한 이미지를 AI가 분석하고 상황을 설명(Captioning)해 줍니다.")

    with gr.Row():
        with gr.Column():
            gr.Markdown("###  이미지 업로드")
            image_input = gr.Image(type="filepath", label="분석할 이미지를 올려주세요")
            analyze_btn = gr.Button(" AI에게 이미지 분석 요청하기", variant="primary")

        with gr.Column():
            gr.Markdown("###  AI의 이미지 분석 결과 (영어)")
            output_text = gr.Textbox(label="Generated Caption", lines=4, interactive=False)

    analyze_btn.click(
        fn=analyze_image_with_hf,
        inputs=image_input,
        outputs=output_text
    )

    app.launch(debug=True)
```
