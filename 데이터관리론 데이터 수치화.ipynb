```python
import pandas as pd
from IPython.display import display
```

```python
url = 'https://raw.githubusercontent.com/ancestor9/data/main/customers.csv'
df = pd.read_csv(url)
```

```python
print("[전처리 전] 원본 데이터 5줄 미리보기:")
display(df.head())
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
```

```python
df_processed = pd.get_dummies(df, columns=categorical_columns, drop_first=False, dtype=int)
```

```python
print(" [전처리 후] 모든 데이터가 수치화된 데이터셋 5줄 미리보기:")
display(df_processed.head())
```
