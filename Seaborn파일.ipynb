```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
population_tips = sns.load_dataset('tips')
print(f"모집단 전체 크기: {len(population_tips)}")
```

```python
sample_simple = population_tips.sample(n=50, random_state=42)

print("1. 단순 무작위 추출 결과 크기:", len(sample_simple))
display(sample_simple.head(3))
```

```python
sample_stratified = population_tips.groupby('day', group_keys=False).apply(
    lambda x: x.sample(frac=0.2, random_state=42)
)

print("\n2. 층화 추출 결과 크기:", len(sample_stratified))
print("[모집단 요일 비율]\n", population_tips['day'].value_counts(normalize=True))
print("[표본 요일 비율]\n", sample_stratified['day'].value_counts(normalize=True))
```

```python
k = 5

start_index = np.random.randint(0, k)

sample_systematic = population_tips.iloc[start_index::k]

print(f"\n3. 계통 추출 결과 크기 (간격={k}, 시작점={start_index}):", len(sample_systematic))
display(sample_systematic.head(12))
```
