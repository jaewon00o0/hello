```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 1. 1페이지부터 10페이지까지 모든 quotes 추출
```

```python
quotes_list = []
for page in range(1, 11):
    url = f"http://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    quotes = soup.find_all('span', class_='text')
    for quote in quotes:
        clean_text = quote.text.strip('“”')
        quotes_list.append(clean_text)
```

```python
# 2. CountVectorizer를 사용하여 빈도 분석 (가장 출현빈도가 높은 단어 5개)
```

```python
vectorizer_freq = CountVectorizer(stop_words='english')
X_freq = vectorizer_freq.fit_transform(quotes_list)
```

```python
word_counts = np.array(X_freq.sum(axis=0))[0]
words = vectorizer_freq.get_feature_names_out()
```

```python
df_word_freq = pd.DataFrame({'word': words, 'count': word_counts})
top5_words = df_word_freq.sort_values(by='count', ascending=False).head(5)
```

```python
print(top5_words.to_string(index=False))
print("\n")
```

```python
# 3. 나타나면 1, 아니면 0으로 Encoding한 데이터프레임 제시
```

```python
vectorizer_bin = CountVectorizer(binary=True)
X_bin = vectorizer_bin.fit_transform(quotes_list)
```

```python
df_quotes = pd.DataFrame(X_bin.toarray(), columns=vectorizer_bin.get_feature_names_out())
```
